```python
def binning(data):
    maximum_value=max(data)
    minimum_value=min(data)
    print(maximum_value,minimum_value)
    #child,teen,adult,old
    child=[]
    teen=[]
    adult=[]
    old=[]
    #for i in range(0,len(data))
    for i in data:
        if i<=12:
            child.append(i)#insert value in your list
        elif i>=12 and i<=20:
            teen.append(i)
        elif i>=20 and i<=45:
            adult.append(i)
        elif i>=45:
            old.append(i)
    print("bin of children=",child)
    print("bin of teen=",teen)
    print("bin of adult=",adult)
    print("bin of old=",old)

age=[10,1,20,30,15,18,80]
binning(age)
```
